## Class balancing using oversampling

```python
import json
from collections import Counter
import random

# Load data from JSON file
with open('/content/drive/MyDrive/DailySales_5PMS/Dataset/Dailysales_5PMS_JSON.json', 'r') as file:
    data = json.load(file)

# Extract category counts
category_counts = Counter(record['label'] for record in data)

# Determine the maximum number of occurrences among all categories

max_occurrences = max(category_counts.values())

# Create a dictionary to store upsampled data for each category
upsampled_data = {}

# Upsample data for each category with fewer occurrences
for category, count in category_counts.items():
    # Skip categories with the maximum number of occurrences
    if count == max_occurrences:
        upsampled_data[category] = [record for record in data if record['label'] == category]
    else:
        # Calculate the number of samples needed to reach the maximum occurrences
        num_samples = max_occurrences - count
        # Sample with replacement from existing data for the category
        upsampled_data[category] = random.choices([record for record in data if record['label'] == category], k=num_samples)

# Combine upsampled data from all categories into a new dataset
upsampled_dataset = []
for records in upsampled_data.values():
    upsampled_dataset.extend(records)

# Combine original data with upsampled data
balanced_dataset = data + upsampled_dataset

# Optionally, save balanced data to a new JSON file
with open('/content/drive/MyDrive/DailySales_5PMS/Dataset/Dailysales_5PMS_balanced_data.json', 'w') as file:
    json.dump(balanced_dataset, file, indent=4)
```

```python
import json
from sklearn.model_selection import train_test_split

# Load data from JSON file
with open('/content/drive/MyDrive/DailySales_5PMS/Dataset/Dailysales_5PMS_balanced_data.json', 'r') as file:
    data = json.load(file)

# Extract features (X) and labels (y) from the loaded data
X = [record['text'] for record in data]
y = [record['label'] for record in data]

# Split the data into training and test sets with stratification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Combine training data into a dictionary
train_data = [{'text': text, 'label': label} for text, label in zip(X_train, y_train)]

# Combine test data into a dictionary
test_data = [{'text': text, 'label': label} for text, label in zip(X_test, y_test)]

# Save training and test data to separate JSON files
with open('/content/drive/MyDrive/DailySales_5PMS/Train_Test/train_data.json', 'w') as file:
    json.dump(train_data, file)

with open('/content/drive/MyDrive/DailySales_5PMS/Train_Test/validation_data.json', 'w') as file:
    json.dump(test_data, file)
```

# Inference:

```python
import spacy
import pandas as pd

# Load the trained spaCy text classification model
nlp = spacy.load('/content/drive/MyDrive/DailySales_5PMS/model/model-best')

# Load data from CSV file
df = pd.read_csv('/content/drive/MyDrive/DailySales_5PMS/Dataset/Dailysales_5PMS.csv')

# Convert NaN values in 'Text' column to empty strings
df['text'] = df['text'].fillna('')

# Extract evaluation_texts and true_labels from the CSV file
evaluation_texts = df['text'].astype(str).tolist()  # Convert to strings and fetch all records from 'Text' column
true_labels = df['label'].tolist()   # Fetch corresponding labels from 'Label' column

# Process evaluation data
docs = list(nlp.pipe(evaluation_texts))

# Predict labels and extract the most probable label for each record
predicted_labels = []
for doc in docs:
    # Get the label with the highest probability
    most_probable_label = max(doc.cats, key=doc.cats.get)
    predicted_labels.append(most_probable_label)

# Print the predicted labels alongside the true labels
for predicted_label, true_label in zip(predicted_labels, true_labels):
    print("Predicted label:", predicted_label, "vs True label:", true_label)
```

```python
df_table = pd.DataFrame({'Text': evaluation_texts, 'Predicted': predicted_labels, 'True': true_labels})

# Display the table
display(df_table)

df_table.to_csv('/content/drive/MyDrive/DailySales_5PMS/model/5PMS_Trained_data_report.csv')
```
